# Free text analysis
Analysis of comments that were custom entered by the respondents.

This notebook assumes that you have already run the `multiple_answer_cleanup.ipynb` notebook.

```python
import numpy as np
import pandas as pd
from IPython.display import display, Markdown, Latex
df = pd.read_csv('responses_complete.csv', index_col=None, parse_dates=['date'])

#don't truncate results
pd.set_option('display.max_rows', None) # display all rows
pd.set_option('display.max_colwidth', -1) # display full text of columns
```

```python
# a function that calculates how many responses were given in a specific column
def get_num_respondents(field, df):
    # get the number of responses in this field
    num_respondents = df[pd.notnull(df[field])].shape[0]
    return num_respondents
```

## Complaints about specific streets and areas

```python
# make a new column with combined comments
df['comments'] = ' *** Feelings: ' + df['feelings'].fillna('none') + ' *** Problem areas: ' + df['problem_areas'].fillna('none') + ' *** Suggestions: ' + df['suggested_improvements'].fillna('none') + ' *** Additional comments: ' + df['additional_comments'].fillna('none') + ' *** Business owner comments: ' + df['business_additional_comments'].fillna('none') + ' *** Final comments: ' + df['final_comments'].fillna('none')

# fields that correspond with specific street complaints
tag_fields = list(df.filter(regex=(".*_tags_.*")).columns)

# show all the comments related to each tag
for field in tag_fields:
    df2 = df[df[field] == 1]
    
    display(Markdown('# Comments about {} ({} total)'.format(field.upper(), get_num_respondents(field, df2))))
    for row in df2['comments']:
        row = row.replace('***', '\n*')
        display(Markdown(row))
        display(Markdown('---'))

    
```
